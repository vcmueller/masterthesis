# Prepare Roles Dataset

Pre-processing steps to prepare the dataset that contains all roles for creating the Word Embedding model and evaluating the system through a survey.

Source dataset: Skill2vec_50k

```python
# Import Libraries
import pandas as pd
```

```python
# Read the Skill2Vec dataset
df_roles = pd.read_csv(filepath_or_buffer="../0-rawdata/skill2vec_50k.csv", sep=",", encoding="latin1", header = None)

# Fill all nulls
df_roles = df_roles.fillna('')

# Join all skills in a same skill column
df_roles['skills'] = df_roles.drop(columns=[0]).apply(lambda x: ';'.join(x.astype(str)), axis=1)
df_roles["id"] = df_roles[0]

# Select only required columns
df_roles = df_roles[['id', 'skills']]
df_roles
```

```python
# Save data frame to CSV
df_roles.to_csv("../2-data/allroles.csv")
```

```python
# Get list of all skills available to be selected - Source: Stack Overflow Survey questions
df_skills = pd.read_csv(filepath_or_buffer="../0-rawdata/skill-list.csv", sep=",", encoding="latin1")

# Get all unique skills
skills = df_skills['skill']

# Account for synonyms
synonym_skill = df_skills['synonym']
def find_index(arr, val):
    for i in range(len(arr)):
        if arr[i] == val:
            return i
    return -1

# Check if any of the mapped skills are listed in the role
count = 0
for skill in skills:
    count = count + df_roles["skills"].apply(lambda x: 1 if skill.lower() in [item.lower() for item in x.split(';')] or synonym_skill[find_index(skills, skill)] in [item.lower() for item in x.split(';')] else 0)
    df_roles['KeepRow'] = count

# Get only roles that have mapped skills
df_roles = df_roles.loc[(df_roles['KeepRow'] > 0)]

# Get only required columns
df_roles = df_roles[["id", "skills"]]
df_roles
```

```python
# Save data frame to CSV
df_roles.to_csv("../2-data/ITroles.csv")
```
