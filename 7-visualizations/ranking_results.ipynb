# Get Results generated from the Survey Part 2


## Import Libraries

```python
import pandas as pd
import numpy as np
```

# Results from Ranking - For each IT Professional

```python

df_results = pd.read_csv(filepath_or_buffer=f"../6-results/ranking_ITProfessional_1.csv", sep=",", encoding="latin1")
df_results['person'] = 1
df_results = df_results[['person','id', 'rank-model1', 'rank-model2', 'rank-model3']]
df = df_results

for person in range(2, 21):

    df_results = pd.read_csv(filepath_or_buffer=f"../6-results/ranking_ITProfessional_{person}.csv", sep=",", encoding="latin1")
    df_results['person'] = person
    df_results = df_results[['person','id', 'rank-model1', 'rank-model2', 'rank-model3']]
    df = df.append(df_results, ignore_index=True)


df.to_excel(f'ranking_results_people.xlsx', index=False)
df
```

# Results from Ranking - For each Role

```python

#df_results = pd.read_csv(filepath_or_buffer=f"../6-results/ranking_role_11.csv", sep=",", encoding="latin1")
#df_results['role'] = 11
#df_results = df_results[['role','id', 'rank-model1', 'rank-model2', 'rank-model3']]
df = pd.DataFrame()

for manager in range(1, 5):
    for role in range(1, 6):
        
        df_results = pd.read_csv(filepath_or_buffer=f"../6-results/ranking_role_{manager}{role}.csv", sep=",", encoding="latin1")
        df_results['role'] = f'{manager}{role}'
        df_results = df_results[['role','id', 'rank-model1', 'rank-model2', 'rank-model3']]
        df = df.append(df_results, ignore_index=True)


df.to_excel(f'ranking_results_roles.xlsx', index=False)
df
```
